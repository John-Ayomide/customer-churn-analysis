```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')

print("Data loaded")
print("Shape:", df.shape)
```

```python
print("TotalCharges data type:", df['TotalCharges'].dtype)
print("\nSample values:")
print(df['TotalCharges'].head(10))

#checking datatype of the column "TotalCharges" which is a str instead of int
```

```python
# Convert TotalCharges from text to a number
# errors='coerce' means: if a value can't be converted, 
# turn it into NaN (blank) instead of crashing

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Check it worked
print("New data type:", df['TotalCharges'].dtype)
print("\nAny missing values created?", df['TotalCharges'].isnull().sum())
```

```python
# looking at those 11 problem rows
# What do they have in common?

missing = df[df['TotalCharges'].isnull()]
print("Number of missing rows:", len(missing))
print("\nTheir tenure values:")
print(missing['tenure'].value_counts())
```

```python
# Since these customers have 0 tenure and no charges yet
# it makes sense to fill the missing TotalCharges with 0
# rather than deleting those customers entirely

df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Verify no missing values remain
print("Missing TotalCharges after fix:", df['TotalCharges'].isnull().sum())
print("Expected: 0")
```

```python
# Currently Churn says "Yes" or "No"
# Machine learning models need numbers not words
# So we convert: Yes = 1, No = 0

df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

print("Churn values after fix:")
print(df['Churn'].value_counts())
print("\nData type:", df['Churn'].dtype)
```
